```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/YashBoda/Salary_Prediction/refs/heads/main/Salary_Data.csv')
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
Experience_train,Experience_test,Salary_train,Salary_test=train_test_split(df[['Experience']],df.Salary,test_size=0.2)
```

```python
Experience_train
```

```python
Experience_test
```

```python
Salary_train
```

```python
Salary_test
```

```python
%matplotlib inline
plt.xlabel('EXPERIENCE[yrs]')
plt.ylabel('SALARY[rs]')
plt.scatter(df.Experience,df.Salary,color='green',marker='*')
plt.title("SCATTERD PLOT OF TOTAL DATA IN 2D PLANE OF EXPERIENCE vs SALARY")
```

```python
%matplotlib inline
plt.xlabel('EXPERIENCE[yrs]')
plt.ylabel('SALARY[rs]')
plt.scatter(Experience_train,Salary_train,color='green',marker='*')
plt.title("SCATTERD PLOT OF TRAINING DATA IN 2D PLANE OF EXPERIENCE vs SALARY")
```

```python
reg=linear_model.LinearRegression()
reg.fit(Experience_train,Salary_train)
```

```python
%matplotlib inline
plt.xlabel('EXPERIENCE[yrs]')
plt.ylabel('SALARY[rs]')
plt.scatter(Experience_train.values,Salary_train.values,color='green',marker='+')
plt.plot(Experience_train.values,reg.predict(Experience_train),color='black')
plt.title("CONTINOUS PLOT OF TRAINING DATA THROUGH OLS REGRESSION")
```
